### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
# Definición de tiendas y nombres
tiendas = [tienda, tienda2, tienda3, tienda4]
tienda_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Colores personalizados para los gráficos de pastel
colores_tiendas = [
    ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],  # Colores para Tienda 1
    ['#c2c2f0', '#ffb3e6', '#c2f0c2', '#ffccf2'],  # Colores para Tienda 2
    ['#ffb3b3', '#b3b3ff', '#b3ffb3', '#ffb3cc'],  # Colores para Tienda 3
    ['#ffcc99', '#c2c2c2', '#ffccff', '#99ccff']   # Colores para Tienda 4
]

# Análisis y visualización
for nombre_tienda, tienda, colores in zip(tienda_nombres, tiendas, colores_tiendas):
    print(f"\n{nombre_tienda}:\n")

    # Agrupación y agregación de datos
    ventasCategoria = tienda.groupby('Categoría del Producto').agg(
        Cantidad_Ventas=('Precio', 'count'),
        Recaudado_Total=('Precio', 'sum')
    ).reset_index().sort_values(by='Cantidad_Ventas', ascending=False)

    print(ventasCategoria)

    # Visualización con gráfico de pastel
    plt.figure(figsize=(6, 6))
    plt.pie(
        ventasCategoria['Cantidad_Ventas'],
        labels=[f"{cat} ({cant})" for cat, cant in zip(ventasCategoria['Categoría del Producto'], ventasCategoria['Cantidad_Ventas'])],
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        colors=colores
    )
    plt.title(f'Distribución de ventas por categoría - {nombre_tienda}')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

print("Si bien la categoría Electrónicos es la segunda más popular, es la primera en recaudación por amplia diferencia.")
```

# 4. Productos más y menos vendidos

```python
def analizar_productos(tiendas_dict, top_n=5):
    """
    Analiza productos más y menos vendidos y genera un cuadro con los resultados.
    """
    for nombre, df in zip(tienda_nombres, tiendas_dict):
        # Verificación y limpieza de datos
        if 'Producto' not in df.columns:
            print(f"⚠️ {nombre}: No tiene columna 'Producto'")
            continue

        # Conteo de ventas por producto
        ventas = df['Producto'].value_counts()

        # Productos TOP y FLOP
        top_productos = ventas.head(top_n)
        flop_productos = ventas.tail(top_n)

        # Crear cuadro de productos más y menos vendidos
        plt.figure(figsize=(8, 4))
        plt.axis('tight')
        plt.axis('off')

        # Datos para la tabla
        tabla_datos = []
        for prod, cantidad in top_productos.items():
            tabla_datos.append([prod, cantidad, 'Top'])
        for prod, cantidad in flop_productos.items():
            tabla_datos.append([prod, cantidad, 'Flop'])

        # Crear la tabla
        plt.table(cellText=tabla_datos, colLabels=['Producto', 'Unidades Vendidas', 'Clasificación'], cellLoc='center', loc='center')
        plt.title(f'Productos Más y Menos Vendidos - {nombre}')
        plt.show()

# Ejecución del análisis
analizar_productos(tiendas)
```
